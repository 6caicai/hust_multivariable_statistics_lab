```python
stock_number = 600925
close_price = np.array([[5.43, 5.46, 5.39, 5.37, 5.4, 5.33, 5.31, 5.34, 5.27, 5.33, 5.27, 5.38, 5.32, 5.47, 5.34, 5.62],
                        [1029, 1028, 1025, 1024, 1023, 1022, 1021, 1018, 1017, 1016, 1015, 1014, 1011, 1010, 1009, 1008]])
close_price = np.flip(close_price)
```

```python
close_price[1].mean(), close_price[1].var()
```

```python
close_price
```

```python
import matplotlib.pyplot as plt

# Separate time and close prices
time = close_price[0]
price = close_price[1]

# Create a figure with two subplots
fig, axs = plt.subplots(2, 1, figsize=(10, 10))

# 1. Histogram of Close Prices
axs[0].hist(price, bins=10, color='skyblue', edgecolor='black')
axs[0].set_title('Histogram of Close Prices')
axs[0].set_xlabel('Close Price')
axs[0].set_ylabel('Frequency')

# 2. Growth Curve (Line Plot)
axs[1].plot(time, price, marker='o', color='orange', label='Close Price')
axs[1].set_title('Growth Curve of Close Prices Over Time')
axs[1].set_xlabel('Time')
axs[1].set_ylabel('Close Price')
axs[1].legend()

# Show the plots
plt.tight_layout()
plt.show()
```
