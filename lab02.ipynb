```python
#ex3
import numpy as np

# Parameters
mu = 1
sigma = 1
sample_size = 10000

# Generate a large sample
sample = np.random.normal(mu, sigma, sample_size)

# Estimate the coefficient of variation
coefficient_of_variation = np.std(sample) / np.mean(sample)
print(coefficient_of_variation)
```

```python
import matplotlib.pyplot as plt

# Function to estimate coefficient of variation for different sample sizes
def estimate_coefficient_of_variation(sample_sizes, mu, sigma):
    estimates = []
    for size in sample_sizes:
        sample = np.random.normal(mu, sigma, size)
        cv = np.std(sample) / np.mean(sample)
        estimates.append(cv)
    return estimates

# Sample sizes to test
sample_sizes = np.arange(100, 10001, 100)

# Estimate coefficient of variation for different sample sizes
estimates = estimate_coefficient_of_variation(sample_sizes, mu, sigma)

# Plot the estimates
plt.plot(sample_sizes, estimates, label='Estimated Coefficient of Variation')
plt.axhline(y=sigma/mu, color='r', linestyle='--', label='True Coefficient of Variation')
plt.xlabel('Sample Size')
plt.ylabel('Coefficient of Variation')
plt.legend()
plt.show()
```

```python
ticker_list = { "INTC": "Intel",
                "MSFT": "Microsoft",
                "IBM": "IBM",
                "BHP": "BHP",
                "TM": "Toyota",
                "AAPL": "Apple",
                "AMZN": "Amazon",
                "BA": "Boeing",
                "QCOM": "Qualcomm",
                "KO": "Coca-Cola",
                "GOOG": "Google",
                "SNE": "Sony",
                "PTR": "PetroChina"}
```

```python
import pandas as pd
df = pd.read_csv("ticker_data.csv")
df1 = pd.DataFrame()
```

```python
times = len(df)
for i in range(1, times):
    for j in ticker_list:
        price_former = df[j][i-1]
        price_present = df[j][i]
        df1.loc[i, j] = (price_present - price_former) / price_former * 100
        print(f"day {i}'s percentage change of {j} is: {(price_present - price_former)/price_former*100}")
```

```python
fig, axes = plt.subplots(3, 5, figsize=(15, 10))  # 3x4 
axes = axes.flatten()  # 

# 
for idx, ticker in enumerate(df1.columns):
    axes[idx].plot(df1[ticker], label=ticker)
    axes[idx].set_title(f"{ticker} - Stock Price Change")
    axes[idx].set_xlabel("Day")
    axes[idx].set_ylabel("Percentage Change (%)")
    axes[idx].legend()
    axes[idx].grid(True)

# 
plt.tight_layout()

# 
plt.show()
```
